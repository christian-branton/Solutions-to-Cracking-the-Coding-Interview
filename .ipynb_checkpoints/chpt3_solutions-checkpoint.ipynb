```python
class Stack:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)
    
    def printStack(self):
        print(['Bottom']+self.items)

```

3.2) I implement a mins stack (which is an attribute of the stack class) which keeps track of the min in the stack for all values below.  That is, when I push a value onto the stack, if it is less than the top of the min stack, I push the value onto the min stack as well.  If it is more, then I push another copy of the top of the min stack onto the min stack.

When I pop the top value off of the stack, I also pop off the top of the min stack.

This allows $O(1)$ execution for the pop(), push() and GetMin() methods.  However, This is at the expense of an increase in the space complexity of $O(n)$ to hold the min stack

```python
class StackMin(Stack):
    def __init__(self):
        self.items = []
        self.mins = []
        
    def push(self, item):
        self.items.append(item)
        
        if len(self.mins)>0:
            if item < self.mins[-1]:
                self.mins.append(item)
            else:
                self.mins.append(self.mins[-1])     
        else:
            self.mins.append(item)
                
    
    def pop(self):
        self.mins.pop()
        return self.items.pop()
    
    def GetMin(self):
        return self.mins[-1]
    
    
my_stack = StackMin()
my_stack.push(11)
my_stack.push(3)
my_stack.push(7)
my_stack.push(8)
my_stack.push(1)
my_stack.push(9)
my_stack.push(11)
my_stack.push(3)

my_stack.printStack()

print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
```

Note, that it is not necessary to hold redundant copies of the same minimum value in the stack.  For example, if the top of the min stack is 2, and I push on values, 4, 3 and 8 onto the stack, the minimum in the stack still remains 2 and so I do not need to push extra copies of 2 onto the min stack.  This optimization will not change the worse case space complexity, which happens if we add items into the stack in reverse sorted order, but in general should certainly help a lot to reduce space. 

I implement this improvement below.  The changes aren't big.  In the pop() method, I only pop the min stack if the popped value on the main stack = the top of the min stack.  For pushing, I only push a value onto the min stack if the new value is less than the top of the min stack.

```python
class StackMin2(Stack):
    def __init__(self):
        self.items = []
        self.mins = []
        
    def push(self, item):
        self.items.append(item)
        
        if len(self.mins)>0:
            if item < self.mins[-1]:
                self.mins.append(item)   
        else:
            self.mins.append(item)
                
    
    def pop(self):
        if self.items[-1] == self.mins[-1]:
            self.mins.pop()
        return self.items.pop()
    
    def GetMin(self):
        return self.mins[-1]
    
my_stack = StackMin2()
my_stack.push(11)
my_stack.push(3)
my_stack.push(7)
my_stack.push(8)
my_stack.push(1)
my_stack.push(9)
my_stack.push(11)
my_stack.push(3)

my_stack.printStack()

print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
my_stack.pop()
my_stack.printStack()
print(my_stack.GetMin())
```

3.3)  To implement SetofStacks, I need to define a maximum capacity allowed for each stack.  Once this capacity is achieved, I must start a new stack.  I store all the stacks in an array, where the first filled stack is at the 0th index in the array, and the current stack is at the last position in the array.  This will allow me to pop and push in O(1) time.  To implement the push() method, I first check the size of the last stack in the array.  If it is less than capactiy, I push the value onto this stack.  However, if it is at capacity I append a new empty stack to the end of the array and push the value onto the stack.  To pop, I pop off the value on the stack in the last element of the array. If the stack becomes empty after the pop, I pop off that cell in the array, so that the second to last stack now becomes the last stack.

```python
class SetofStacks:
    def __init__(self, capacity):
        self.arr = [[]]
        self.capacity = capacity

    def push(self, item):
        if len(self.arr[-1]) < self.capacity:
            self.arr[-1].append(item)
        else:
            self.arr.append([item])

    def pop(self):
        val = self.arr[-1].pop()
        if len(self.arr[-1]) == 0:
            self.arr.pop()
        return val

    def printStack(self):
        print([["Bottom"] + x for x in self.arr])
        
my_stack = SetofStacks(3)
my_stack.push(11)
my_stack.push(3)
my_stack.push(18)
my_stack.push('dog')
my_stack.push(True)
my_stack.push(False)
my_stack.push(3.14159)
my_stack.push(2.99792458e8)

my_stack.printStack()
my_stack.pop()
my_stack.printStack()
my_stack.pop()
my_stack.pop()
my_stack.pop()
my_stack.printStack()
```

To pop in the middle of the stack, I first need to locate the cell in the array containing the appropriate stack and the appropriate index in that stack.  I can achieve this with the // and % operators respectively.  I then access that value, save it, and pop that element within that stack.  

I must now rearrange the elements in the next few stacks.  Specifically, for the stack ahead, I must remove the bottom value and place it at the top of the previous stack (and I must do this for all subsequent stacks) to respect the capactiy constraint that all stacks up until the last stack in the array have exactly capactiy elements.  After this procedure, it is possible that the last stack has zero elements.  If this is the case I delete the last element in the array.

In this case, since the pop could occur early in the stacks, the time complexity is now O(n) to take care of rearranging the positions of all subsequent elements.

```python
class SetofStacks(SetofStacks):    
    def PopAt(self, i):
        stack_num = i // self.capacity
        index_in_stack = i % self.capacity
        val = self.arr[stack_num][index_in_stack]
        self.arr[stack_num].pop(index_in_stack)
        
        for i in range(stack_num+1, len(self.arr)):
            new = self.arr[i].pop(0)
            self.arr[i-1].append(new)
        
        if len(self.arr[-1]) == 0:
            self.arr.pop()
        
        return val

my_stack = SetofStacks(3)
my_stack.push(11)
my_stack.push(3)
my_stack.push(18)
my_stack.push('dog')
my_stack.push(True)
my_stack.push(False)
my_stack.push(3.14159)
my_stack.push(2.99792458e8)

my_stack.printStack()
my_stack.PopAt(4)
my_stack.printStack()
my_stack.PopAt(0)
my_stack.printStack()
my_stack.PopAt(0)
my_stack.printStack()
my_stack.PopAt(4)
my_stack.printStack()
```

3.4) I can implement a queue with 2 stacks in the following manner:  I have a main stack which stores all of the data I push onto the queue by pushing these values onto the stack.  In a queue, popping off an element pops off the first element added to the queue, while popping an element off a stack pops off the last element added to the stack.  Therefore to replicate this behavior, I transfer all elements in my main stack (via popping) to my helper stack (via pushing), which results in a stack in the opposite order.  I can then pop the helper stack, which pops off the last element placed in the helper stack, which was the first element placed on the main stack.  This therefore replicates the popping behavior of a queue.  To maintain proper order, I must then transfer back all elements (except for the element I popped) from the helper stack to the main stack. 

Note, that unlike a normal queue which pushes in O(n) time and pops in O(1) time, MyQueue will push in O(1) time and pop in O(n) time.

```python
class MyQueue:
    def __init__(self):
        self.stack_main = Stack()
        self.stack_helper = Stack()
    
    def push(self, val):
        self.stack_main.push(val)
     
    def pop(self):
        
        # transfer contents from main to helper
        while not self.stack_main.isEmpty():
            self.stack_helper.push(self.stack_main.pop())
        
        # pop off element in helper (and save to return later), 
        # then transfer elements back to main
        val = self.stack_helper.pop()
        while not self.stack_helper.isEmpty():
            self.stack_main.push(self.stack_helper.pop())
        
        return val
    
    def Print(self):
        print(["Back"]+list(reversed(self.stack_main.items))+["Front"])

my_q = MyQueue()
my_q.push('A')
my_q.push('B')
my_q.push('C')
my_q.push('D')
my_q.push('E')
my_q.push('F')
my_q.Print()

my_q.pop()
my_q.Print()

my_q.pop()
my_q.pop()
my_q.Print()
```

3.5) To sort the stack with only another stack as spare space, I first create an empty stack, stack2.  Then I transfer all elements from the stack to stack2 in a while loop and save the value of the biggest element as well as the size of stack1. I then transfer all elements from stack2 back to the stack (except for the biggest element) by running a for loop with the number of iterations equal to the size of the stack.  I now manually push the bigget element onto stack 2.  Note that now, the size of the stack will be one less since I did not push the biggest value back onto the stack.

I repeat this process in a while loop until the stack is empty, at which point stack2 will contain the sorted stack.

The space complexity is O(n) from the spare stack, and the time complexity is $O(n^2)$ since I have nested loops, where the outer loop runs $n$ times and the the inner loop performs $O(n)$ work.

```python
from math import inf

def SortStack(stack):
    
    stack2 = Stack()

    while not stack.isEmpty():
        ### push values onto stack2 to get biggest value in stack1 and 
        ### get size of stack1 while doing this
        size_stack = 0
        biggest = -inf
        while not stack.isEmpty():
            val = stack.pop()
            if val > biggest:
                biggest = val
            stack2.push(val)
            size_stack += 1

        ### pop off values of stack2 and push onto stack1, except the biggest value,
        ### so that the size of stack1 is one less now.  I use a cnt variable to help 
        ### with duplicates.  Push this biggest value onto stack2 manually.
        cnt = 0
        for _ in range(size_stack):
            val = stack2.pop()
            if val == biggest and cnt == 0:
                cnt +=1
            else:
                stack.push(val)
        stack2.push(biggest)
    
    return stack2

my_stack = Stack()
my_stack.push(11)
my_stack.push(7)
my_stack.push(5)
my_stack.push(11)
my_stack.push(20)
my_stack.push(2)
my_stack.printStack()

sorted_stack = SortStack(my_stack)
sorted_stack.printStack()
```

```python
class AnimalShelter:
    def __init__(self):
        self.animals = [] ## define a queue using a python list
        self.HelperStack = Stack()
    
    def Enqueue(self, name, animal_type):
        self.animals.insert(0, (animal_type, name))
    
    def DequeueOldest(self):
        val = self.animals.pop()
        print('Here is '+val[1]+' the '+val[0])
    
    def DequeueAnimal(self, animal_type):
        val = (None, None)
        
        for _ in range(len(self.animals)):
            val = self.animals.pop()
            if val[0] ==  animal_type:
                break
            else:
                self.HelperStack.push(val)
        
        if val[0] != animal_type:
            print("Sorry, no " + animal_type + 's '+ "currently")
        else:
            print('Here is '+val[1]+' the '+val[0])
        
        while not self.HelperStack.isEmpty():
            self.animals.append(self.HelperStack.pop())
        
    def PrintQueue(self):
        print(["Back"] + self.animals + ["Front"])
        
my_shelter = AnimalShelter()
my_shelter.Enqueue('Woofy', 'dog')
my_shelter.Enqueue('Sparky', 'dog')
my_shelter.Enqueue('Kitty', 'cat')
my_shelter.Enqueue('Whiskers', 'cat')
my_shelter.Enqueue('Bingo', 'dog')
my_shelter.Enqueue('Lassie', 'dog')
my_shelter.Enqueue('Fluffy', 'cat')
my_shelter.PrintQueue()   

my_shelter.DequeueOldest()
my_shelter.PrintQueue()   

my_shelter.DequeueAnimal('cat')
my_shelter.PrintQueue()

my_shelter.DequeueAnimal('dog')
my_shelter.PrintQueue()

my_shelter.DequeueAnimal('dog')
my_shelter.PrintQueue()
```
